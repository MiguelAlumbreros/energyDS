<h1> Nevergrad Testing</h1>

This file implements some basic optimizers on ideal forecast data using nevergrad.
This uses a fixed demand,PV and price dataset. It also uses a very simple battery model as flexible source.

```python
# Import neccesary libraries
import numpy as np
import pandas as pd
import nevergrad as ng
import matplotlib.pyplot as plt
```

<h2> Load Dummy Data</h2>

<b>fixed data for now, with 'ideal' forecasts</b>

Demand data

```python
LoadData = pd.read_csv('Load_data_aggregated_1H.csv',
                       index_col=0, parse_dates=True)
LoadData
```

Price data

```python
Prices = pd.read_csv('Price_Data_aggregated_1H.csv', index_col=0, parse_dates=True)
Prices
```

<h2>Setup Optimizer</h2>

<b>Device Data</b><br>
Very simple device models

```python
batteryCapacity = 200  # kWh

inverterCapacity = 100  # kW
```

<b>Preprocessing</b>

```python
# Helper function for visualizing the data and optionally the proposed battery behaviour
# Helper function for visualizing the data and optionally the proposed battery behaviour
def visualizeResultsCostArbitrage(production, batteryBehaviour=None, startDay=0, endDay=1, injection_gain = False):
    startVal = (startDay)*72
    endVal = (endDay)*72
    xvals = range(startVal, endVal)
    # Plot demand
    plt.plot(xvals, LoadData.demand.values[startVal:endVal], label='Demand')
    # Plot production
    plt.plot(xvals, production, label='Production')
    plt.xlabel('Time [hours]')
    plt.ylabel('Energy [KWh]')
    plt.gca().axhline(linewidth=2, color='black')
    # Plot battery state (if applicable)
    startVal = (startDay)*72
    endVal = (endDay)*72
    if batteryBehaviour is not None:
        plt.plot(
            xvals, batteryBehaviour, label='Battery demand')
        chargeState = np.cumsum(batteryBehaviour)
        plt.plot(xvals, chargeState, label='Battery charge')
        max_bat_charge = np.max(chargeState)
        max_delta = np.max(np.abs(batteryBehaviour))
        print(f'Max battery charge: {max_bat_charge}')
        print(f'Max battery change: {max_delta}')

    if batteryBehaviour is not None:
        total_demand = LoadData.demand.values[startVal:endVal] - production \
            + batteryBehaviour

    else:
        total_demand = LoadData.demand.values[startVal:endVal] - production

    plt.plot(xvals, total_demand, label='Net demand')
    peak = np.max(np.abs(total_demand))
    if not injection_gain:
        total_demand[total_demand < 0] = 0
    cost = np.multiply(total_demand, Prices.price.values[startVal:endVal])
    cost = cost.reshape((-1, 72))
    cost = np.sum(cost, axis=1)
    price = np.sum(cost)
    print(f'Total cost: {price}')       
    print(f'Max net demand: {peak}')
    plt.legend()
    plt.show()    

    return price
```

<b>Define Cost functions</b>

```python
def cost_function_arbitrage(*args, **kwargs):
    individual = kwargs['battery_behavior']*inverterCapacity
    cumsum =  np.cumsum(individual)
    violations_battery = (np.sum(np.where(cumsum>batteryCapacity)) + np.sum(np.where(cumsum<0))) * 10000
    
    #The demand minus the solar production together with the usage of the battery may not be below zero
    total_usage = demand+individual 
    #violations_demand = np.sum(np.where(total_usage<0)) * .01
    total_usage[total_usage < 0] = 0 # Make sure no money is returned!
    
    return (np.dot(price,total_usage) + violations_battery)

def cost_function_arbitrage_mult(*args, **kwargs):
    return cost_function_arbitrage(*args, **kwargs)*10000
```

<h2>Define optimisation problem and solve</h2>

No multiplier added to the cost function:

```python
budget = 10000
day = 9
price = Prices[day*72:(day+1)*72].values[:,0]
demand = LoadData[day*72:(day+1)*72].values[:,0]

parameters = ng.p.Instrumentation(battery_behavior=ng.p.Array(init= np.zeros(len(demand)), mutable_sigma=False)
                                              .set_bounds(-100.0,100.0)
                                              .set_mutation(200.0))

# Define optimizer algorithm, computational budget and dimensionality of vector to be optimized
optimizer = ng.optimizers.Shiwa(parametrization=parameters, budget=budget)
# Add constraints (can also be expanded to include additional constraints such as depth of discharge, cycling constraints etc.)


# Execute optimizer

recommendation = optimizer.minimize(cost_function_arbitrage)  # best value

print('The optimized cost is: ' + str(cost_function_arbitrage(*recommendation.args, **recommendation.kwargs)))
```

```python
production_loc = np.zeros(72)

visualizeResultsCostArbitrage(production_loc,recommendation.kwargs['battery_behavior']*inverterCapacity, day, day+1, False)
```
