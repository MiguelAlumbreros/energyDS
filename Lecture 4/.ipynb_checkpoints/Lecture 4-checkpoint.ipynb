# Lecture 4: Optimization
In this lecture, we take a look at some ways that can be used to optimize energy demand and production in smart grids. A number of different objective functions can be used in this optimization, depending on the context. 

In this lecture, we focus on two commonly encountered optimization problems. The methodology used here - genetic algorithms - is extremely flexible as it belongs to the class of derivative free optimizers. Genetic algorithms have been used to solve just about every optimization problem, ranging from resource routing to training neural networks. However, their flexibility comes at a cost: in many situations, it makes sense to utilize a different optimization method which can make use of the special structure in the problem. In the first section of this lecture, we discuss some of these distinctions that arise in practice.

## Classifying optimization problems
Optimization problems can be classified in many different ways. Some of the more interesting ones include:

### Constrained vs. unconstrained optimization
An unconstrained optimization problem does not impose any constraints on the decision variables. Training a linear model, as we did in the previous lecture, is an example of unconstrained optimization. A constrained optimization problem, on the other hand, is typically formulated with some constraints on the values the decision variables can take (e.g. non-negativity constraints, binary or integer value constraints etc.). Optimization problems in power systems generally involve constraints, stemming either from hardware or user requirements.

### Convex vs. nonconvex optimization
Tyrrell Rockafellar once wrote that _the great watershed in optimization is not between linear and nonlinear problems, but between convex and nonconvex problems_. While convex problems can be solved efficiently up to very large problem sizes, this does not hold for nonconvex problems. A non-convex optimization problem is any problem where the objective or any of the constraints are non-convex, which means there can be multiple feasible regions. While linear programming is an example of a convex optimization problem, large nonconvex problems are typically solved either by convexification (e.g. linearization of constraints etc.) or metaheuristics (such as genetic algorithms). Gradient descent, commonly used to train neural networks, is another example. These techniques generally do not guarantee the optimal solution.

### Centralized vs. distributed optimization
Historically, power systems were treated as monolithic centrally planned systems where the generation capacity of a limited number of electricity producers was modulated according to their operational costs and prevailing grid codes. However, with increasing proliferation of distributed energy sources, this paradigm no longer holds true. Therefore, in many cases it is important to solve the optimization problem in a distributed fashion, i.e. at each node, rather than centrally. This has important implications for the amount of data and information that needs to be transferred as well as the quality of the solution that can be obtained at the end.

### Deterministic vs. stochastic optimization
Typically, optimization is carried out with forecasts for the future. Where there is uncertainty associated with these forecasts (e.g. in the case of electricity demand and supply), it is important to take this information into consideration during the optimization process. This is the difference between deterministic and stochastic optimization. Here we use stochastic optimization to refer to the large family of algorithms which enable this, including robust optimization, stochastic programming etc.

### Single-objective vs. multi-objective optimization
Finally, while many optimization problems require a single objective to be optimized, often times we are interested in optimizing for multiple objectives. In some instances, it is possible to combine these different objectives into a single objective function (e.g. by computing some sort of weighted average, and then using that as an objective function), but this is not always desirable. In this case, we can consider the Pareto front of solutions, which reflect all possible non-dominated solutions, i.e. picking one solution over the other will reduce the objective function value in some way. In this lecture, we will cover two different objective functions, but combining them will be left to the reader.


## Formulating and solving an optimization problem

### Elements of an optimization problem
In general, a constrained optimization problem consists of three key elements:
1. Variables: these can either be decision variables i.e. those that can be manipulated in the optimization process or defined variables, i.e. those which cannot be modified.
2. Constraints: these define any constraints that are imposed on the values the decision variables may or may not take.
3. Objective function: this specifies the metric we would like to optimize.

### Examples

Some common examples in smart grids include optimizing towards the following different objective functions:
1. Minimizing grid electricity costs (using a battery-inverter system)
2. Minimizing grid electricity costs (using a battery-inverter system) when a solar panel is installed as well
3. Minimizing peak demand 
4. Minimizing energy demand and peak demand at the same time

## Minimizing grid electricity costs

In the first part of this lecture, we will take a look at minimizing the electricity usage prices based on cost arbitrage (buy-low-sell-high) with an electrical battery.

### Defining demand and electricity prices

We start off with some imports and other boilerplate as usual.

```python
import numpy as np
from deap import algorithms, base, creator, tools
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 15})
```

First, let's define electricity prices. The price oscillates between maximum and minimum price multiple times in the day for demonstration purposes (in reality, it is straight forward to replace these prices with actual spot market prices - however then calculating the optimal cost savings is not straight forward at all).

```python
price = np.array([10,10,10,20,20,20,10,10,10,20,20,20,10,10,10,20,20,20,10,10,10,20,20,20])

plt.plot(price)
plt.grid(True)
plt.xlabel('Time [Hours]')
plt.ylabel('Electricity cost [$]')
plt.show()
```

Now, let's also define the base demand! 

Note: in practice your base demand will obviously be from the building or cluster of buildings you are analyzing.

```python
demand = (np.array([22,20,25,22,22,26,25,32,44,43,46,36,40,38,45,44,47,38,31,26,22,25,22,24])**3)/1000

plt.plot(demand)
plt.grid(True)
plt.xlabel('Time [Hours]')
plt.ylabel('Electricity demand [kWh]')
plt.show()
```

Finally, let us define the battery and grid characteristics which form part of the environment and cannot be modified by the optimization framework.

```python
batteryCapacity = 75 #kWh
inverterCapacity = 40 #kW
peakPowerLimit = 80 # kW
```

### Setting up the framework
Instead of directly starting off with formulating the optimization problem, it is instructive to consider a simpler approach. A first such attempt can be to simply randomly sample a large number of potential solutions and choose the best one. The idea here is that it is extremely cheap to create and evaluate a solution, while it can be computationally challenging to find the optimal one. There are numerous ways to do this of course, but here we use the DEAP library, a versatile tool that also allows us to perform optimization in a more sophisticated manner.

To start off, we create a new class FitnessMin, inheriting from base class. This is necessary to evaluate the objective function for a given solution. The negative sign means that we are creating a minimization problem.

```python
creator.create("Fitness", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.Fitness)
```

Next, we create a toolbox that will eventually contain any characteristics we wish our solutions to possess.

```python
toolbox = base.Toolbox()
```

We need to register an attribute generator in the toolbox which can be any random integer between +/- battery inverter rated power. This imposes the constraint that our battery-inverter system cannot draw more power than it is physically allowed to.

```python
# Attribute generator
toolbox.register("attr_bool", np.random.randint, -inverterCapacity, inverterCapacity+1)
```

After registering the attributes, we also want to create individual solutions (and also a population of such solutions).

```python
# Structure initializer
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, 24)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

We can also use the toolbox to create an individual solution and visualize what it contains.

```python
np.array(toolbox.individual())
```

Note that so far we have only created a single solution, where every element is the charging or discharging power of the inverter. We could have easily done the same by simply creating a numpy array with random integers. However, in the following, we will explore the versatility of the DEAP library to find optimal (or near-optimal) solutions.

The first of these is to create an entire population of solutions with a single line of code as follows:

```python
np.array(toolbox.population(10))
```

At this point, we need to define the cost function that we wish to minimize. As explained at the beginning of the section, this cost function should reflect the price of electricity and the electricity demand based on the battery usage. We assume, for now, that the excess electricity stored in the battery can be used either in the building or sold back to the grid. We also assume that the battery has 100% efficiency (wishful thinking, but simplifies things to start off).

Finally, we place a very large penalty on battery violations in the cost function. A battery violation is when the battery is charged beyond its capacity (or alternatively discharged below 0 kWh). As this is a constrained optimization problem, it would make sense to formulate these as constraints (which is what you would do in convex optimization formulations). However, in many population based (or derivative free algorithms), it is commonplace to include the constraint in the objective function (with a high cost). This is a relaxation of the constraint. 

You can theoretically also add the battery capacity (or any other systemic limitation) as a constraint.

```python
def cost_function_arbitrage(individual):
    cumsum = np.cumsum(individual)
    violations_battery = (np.sum(np.where(cumsum>batteryCapacity)) + np.sum(np.where(cumsum<0))) * 10000
    return np.sum(price*individual) + violations_battery,
```

```python
toolbox.register("evaluate", cost_function_arbitrage)
```

Make sure you understand the cost function before going forward.

### Random guessing

Now we have all the tools to create a population with x different random solutions and evaluate them. Then,  we can select the best one of the lot!

```python
numSolutions = 10
pop = toolbox.population(n=numSolutions)
cf = np.zeros(numSolutions)
mincf = np.zeros((numSolutions-1))
for i in range(numSolutions):
    cf[i] = cost_function_arbitrage(pop[i])[0]
    if i>0:
        mincf[i-1] = np.min(cf[0:i])
```

We can see the best solution the random search has discovered after evaluating the specified number of solutions. Remember that we are interested in minimizing the total energy cost so lower scores are better.

```python
plt.plot(mincf)
plt.ylim(-3000,100)
plt.show()

print(np.min(mincf))
```

Let us define a helper function to visualize the solutions (including the best solution) we have obtained and its comparison with desired behavior.

```python
def visualizeResults(solution):

    plt.plot(solution,'*-',label='Inverter demand')
    plt.plot(demand,label='Building energy demand')
    plt.plot(price,'o--',label='Price [$]')
    plt.grid(True)
    plt.xlabel('Time [h]')
    plt.ylabel('[kWh]')
    plt.legend()
    plt.title('Best solution overlaid with building energy demand and prices')
    plt.show()

    plt.scatter(price, solution)
    plt.xlabel('Price [$]')
    plt.ylabel('Inverter power [kW]')
    plt.grid(True)
    plt.title('Charge/Discharge operation as a function of electricity prices')
    plt.show()
    
    plt.plot(demand + solution)
    plt.xlabel('Time [h]')
    plt.ylabel('Total grid demand')
    plt.grid(True)
    plt.title('Total electricity demand on the grid')
    plt.show()
```

Now, let's visualize the best result we have discovered through random sampling.

```python
minIndex = np.argmin(cf)
visualizeResults(pop[minIndex])
```

### Setting up genetic algorithms
Instead of a brute force search which retains no memory and therefore draws all solutions independently, it is possible to either optimize this problem using linear programming or a 'smart' search based method. These latter include strategies such as simulated annealing, population based methods (such as genetic algorithms, ant colony optimization) and tabu search etc.

Here, we dig deeper into using genetic algorithms to discover the optimal charging and discharging profile of the battery.

Genetic algorithms offer a large number of levers to play around with to discover optimal solutions. As a member of the broader family of algorithms known as metaheuristics, these algorithms are usually not known for finding the optimal solution. However, they are often extremely versatile tools to discover good enough solutions to many practical problems of interest. Furthermore, they are anytime algorithms, which means they do not need to be run till convergence and the search can be stopped at any point in time to retrieve the best solution that has been discovered so far.

#### Initialization
A population of individuals is initialized, often randomly. This population is then evolved in a way that its fitness improves over time. This is a global optimization process which means that, in the short term, the fitness of the population may go down to avoid getting trapped in local optima.

#### Selection 
A selection process is defined to give preference to fitter individuals in the population in the evolutionary process.

#### Crossover
Takes multiple parents and combines them to form a new indiivdual. Examples of the crossover operation include one-point crossover and edge recombination. The idea behind this operation is to combine the best possible solutions so far to enable discovery of even better solutions.

#### Mutation
Introducing new traits into the population, i.e. more diversity. The idea of this operator is to search for better solutions by varying elements of already discovered solutions in a controlled manner.

Both mutation and crossover, combined with selection, ensure that genetic algorithms retain history over time and therefore continuously move towards an improved solution. This is in contrast to the brute force search we encountered earlier, which did not have this iterative element to it.

#### The brains of the operation
At the end of the day, with a lot of these algorithms, you have to ask whether the credit of finding a good solution lies (1) with the algorithm itself, (2) with the programmer designing the algorithms, or (3) with the solution space. Often, in simple Genetic Algorithm approaches, it is the latter two, as the algorithm itself is not 'particularly' smart or clever. However, with increasing compute power, it provides good results, surprisingly often. Furthermore, as the search algorithm remains the same regardless of the objective function, GA can be used to solve even non-convex problems with multiple objective functions.

The first step to explore genetic algorithms for our problem, using in-built DEAP functionality, is therefore to define the mating, mutation and selection processes. For more practical information on using genetic algorithms, check out the DEAP documentation. For a high level but slightly more technical overview, see [here](https://www.youtube.com/watch?v=wiQOsH8C-uQ&t=936s) and [here](https://www.youtube.com/watch?v=kHyNqSnzP8Y). 

An excellent, highly recommended description of evolutionary algorithms is presented in this [visual guide](http://blog.otoro.net/2017/10/29/visual-evolution-strategies/).

For the purpose of this lecture, we define the mating, mutation and selection operators. For more details, refer to the DEAP documentation.

```python
toolbox.register("mate", tools.cxTwoPoint) # crossover
toolbox.register("mutate", tools.mutUniformInt, low=-inverterCapacity, up=inverterCapacity, indpb=0.2) # mutation
toolbox.register("select", tools.selTournament, tournsize=3) # selection 
```

Initialize a population of 10,000 offspring (solutions). These are created randomly according to the rules we specified above. Chances are high that most, if not all, of these solutions will be unfeasible and will therefore have really awful fitness values (as before, with the brute force search).

```python
pop = toolbox.population(n=10000)
```

Now, we can initialize the hall of fame, aka a bunch of solutions which have outperformed all other solutions that we have encountered so far.

```python
hof = tools.HallOfFame(10)
```

And finally, before running the genetic algorithm itself, we would like to create the metrics which allow us to evaluate not just the best solution found so far (which is done through the fitness function), but also the fitness function values of the entire population (as the minimum, maximum and average values; the standard deviation in fitness is interesting as we want there to be enough diversity in our solutions, especially at initialization)

```python
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("avg", np.mean, axis=0)
stats.register("std", np.std, axis=0)
stats.register("min", np.min, axis=0)
stats.register("max", np.max, axis=0)
```

With everything defined, we can set the genetic algorithm in motion now!

```python
# Single objective case
pop, log = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.5, ngen=10, stats=stats, halloffame=hof,  verbose=True)
```

Now, we can finally visualize the best solution that we have discovered so far!

```python
visualizeResults(hof[0])
```

We can also visualize the different solutions in the hall of fame to see if they are all different or similar!

```python
plt.plot(np.array(hof).T,'black',alpha=.3,linewidth=0.5)
plt.plot(np.mean(np.array(hof).T,1),'red',linewidth=3)
plt.boxplot(np.array(hof))
plt.xlabel('Time of day')
plt.ylabel('Inverter demand [kW]')
plt.grid(True)
plt.show()
```

## Extending the objective function

Previously, we have specifically focused on using the battery-inverter system to provide arbitrage using price differences. This has assumed that the building energy demand can go negative, i.e. the controller is allowed to inject electricity back to the grid if it thinks this would be beneficial. Assuming that grid injection is not possible (or remunerated, as is often the case), we can change the cost function appropriately to discourage this quite easily. This gives us a new cost function which we then need to register again.

```python
def cost_function_arbitrage_constrained(individual):
    cumsum = np.cumsum(individual)
    violations_battery = (np.sum(np.where(cumsum>batteryCapacity)) + np.sum(np.where(cumsum<0))) * 10000
    
    tmp = demand+individual
    violations_demand = -np.sum(tmp[np.where(tmp<0)]) * 10000
    
    return np.sum(price*individual) + violations_battery + violations_demand,
```

```python
toolbox.register("evaluate", cost_function_arbitrage_constrained)
```

Reinitialize both the population and the hall of fame.

```python
pop = toolbox.population(n=10000)
hof = tools.HallOfFame(10)
```

```python
pop, log = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.5, ngen=10, stats=stats, halloffame=hof,  verbose=True)
```

```python
visualizeResults(hof[0])
```

If the optimizer has done its job, we should see that the total building energy demand no longer goes below zero, while still reducing energy costs for the end user.
